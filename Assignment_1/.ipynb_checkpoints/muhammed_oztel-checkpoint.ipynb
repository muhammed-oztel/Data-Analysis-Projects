# Imports

```python
from datascience import *
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

# 1. Data Reading

```python
# Here we read data from file using read table method
background = Table.read_table('Background.csv')
grades = Table.read_table('Grades.csv')
```

# 2. Columns & Rows reporting

```python
# to display the background table, it has 6 columns and 1000 rows
# column names are ID, gender, race/ethnicity,parental level of education,lunch,test preparation course
background
```

```python
# to display grades tables, it has 4 columns and 1000 rows Column names are ID, math score, reading, score, writing score
grades
```

# 3. Average grade

```python
# adding new column Average Score based on the 3 scores
grades = grades.with_column('Average Score', (grades.column('math score')*40 + (grades.column('reading score')*30)+
                                              (grades.column('writing score')*30))/100)
grades
```

# 4. Sorting

```python
# display average scores higher and equal to 95 in max min order
grades.sort('Average Score', descending = True).where('Average Score', are.above_or_equal_to(95))
```

# 5. Scatter Plots

```python
# to plot math and reading scores relation with scatter and also there is a good relation between the scores
grades.scatter('math score','reading score')
```

```python
# to plot math and wring scores relation with scatter and also there is a pretty nice relation between the scores
grades.scatter('math score','writing score')
```

```python
# to plot reading and wring scores relation with scatter and also there is a more strong relation between the scores
grades.scatter('reading score','writing score')
```

We can say that there is a relation between the scores for each graph.But writing and reading realation is much stronger.

# 6. Gender Analysis

```python
# grouping data based on gender to display the average of each score type
background.join('ID', grades).drop([0,2,3,4,5]).group('gender',np.average)
```

# 7. Letter Grade

```python
# this function is to set letter grade with respect to input grade
def letter_grade(grade):
    if grade>=90:return 'A'
    elif 80<=grade<90:return 'B'
    elif 70<=grade<80:return 'C'
    elif 60<=grade<70:return 'D'
    elif grade<60:return 'F'
# adding new column using the function created above to set letter based on grades, using apply mehtod
grades = grades.with_column('Letter Grade',grades.apply(letter_grade,'Average Score'))
grades
```

```python
# plotting to see distrubition of letter grades
grades.drop([0,1,2,3,4]).group('Letter Grade').plot('Letter Grade')
```

As we can see the letter grade with the highest frequency is F 

# 8. Race/Ethnic Group Analysis

```python
# groping with race and letters 
background.join('ID', grades).group(['race/ethnicity', 'Letter Grade']).where('Letter Grade',
                                                                              'F').sort('count',descending= True).take(0)
```

As we can see group C has the highest number of failures

# 11. Education Level

```python
# here we look for the grades of A or B and the parental education level
total = background.join('ID',grades)
ab = total.where('Letter Grade', are.contained_in('AB'))
deg = total.where('parental level of education', are.not_containing('high school'))
hgh = total.where('parental level of education', are.containing('high school'))
ab_deg=ab.where('parental level of education', are.not_containing('high school'))
ab_hgh = ab.where('parental level of education', are.containing('high school'))
per_deg = ab_deg.num_rows / deg.num_rows *100
per_hgh  =ab_hgh.num_rows/ hgh.num_rows *100
per_deg, per_hgh
```

As we can see 24 percent had an education higher than high school
and approximately 12.67 percent did not have an education higher than high school
